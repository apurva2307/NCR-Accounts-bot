```python
def convert_to_datetime(df):
    result = []
    for index, row in df.iterrows():
        try:
            result.append(pandas.to_datetime(row["W.E.F."], format='%d-%m-%Y'))
        except:
            result.append(0)
    return result
data['W.E.F.'] = convert_to_datetime(data)

```
